```python
# import statement

import os
import sys
from collections import namedtuple
import numpy 
import pandas 
import matplotlib.pyplot as plt
import pylab
```

```python
# Define the file path here
file_path = 'sample_data/San Francisco FY14-15.csv'

# read the CSV file using pandas
data = pandas.read_csv(file_path) # returned as a data frame

# display the original data's head
data.head()
```

```python
# Method that processes the read dataframe from CSV file
def denoise_DataFrame(data):
    # get rid of the unnamed column
    data.columns = data.columns.str.replace('^Unnamed:.*', '')    
    data = data[[col for col in data.columns if col]]

    # replace the ' ' to '_' in the column
    data.columns = data.columns.map(lambda x: x.replace(' ', '_') if isinstance(x, (bytes,str)) else x)
    return data
```

```python
data = denoise_DataFrame(data)

# display the processed data's head
data.head()
```

```python
# function that will trim the money sign and whitespace
def processAmount(s):
    return str(s).replace('$', '').strip()
```

```python
# Create the Purchase model as namedtuple
Purchase = namedtuple('Purchase', 'POST_DATE VENDOR DESCRIPTION QTY INVOICE_AMT')
purchase_list = []

i = 0
for row in data.itertuples():
    new_purchase = Purchase(row.POST_DATE, row.VENDOR, row.DESCRIPTION, row.QTY, processAmount(row.INVOICE_AMT))
    purchase_list.append(new_purchase)
```
